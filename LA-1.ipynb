```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["WorkWeekHrs"])

plt.pyplot.xlabel("WorkWeekHrs")
plt.pyplot.ylabel("count")
plt.pyplot.title("WorkWeekHrs bins")
```

```python
bins = np.linspace(min(df["WorkWeekHrs"]), max(df["WorkWeekHrs"]), 4)
bins
```

```python
group_names = ['Low', 'Normal', 'High']
```

```python
df['WorkWeekHrs-binned'] = pd.cut(df['WorkWeekHrs'], bins, labels=group_names, include_lowest=True )
df[['WorkWeekHrs','WorkWeekHrs-binned']].head()
```
